## GeoJSON

```python
def GeoJSON(data):
    bundle = {}
    bundle['application/geo+json'] = data
    bundle['text/plain'] = data
    display(bundle, raw=True)
    
GeoJSON({
  "type": "Feature",
  "geometry": {
    "type": "Point",
    "coordinates": [125.6, 10.1]
  },
  "properties": {
    "name": "Dinagat Islands"
  }
})
```
